Find if there is a significant difference in the diameter of the cutlet between two units.

To solve this problem we will conduct a 2 Tailed Test on two independent samples with numberical data. For this we will check whether the mean of both samples are different.

```python
plt.figure(figsize=(8, 6))

labels = ['Unit A', 'Unit B']

sns.histplot(df['Unit A'], color='blue')
sns.histplot(df['Unit B'], color='orange')
sns.kdeplot(df['Unit A'], color='blue')
sns.kdeplot(df['Unit B'], color='orange')

plt.legend(['Unit A', 'Unit B'])
plt.show()
```

Checking Normal Distribution

```python
plt.figure(figsize = (5,4))
sm.qqplot(df["Unit A"], line = 'q')
plt.title('Unit A')
sm.qqplot(df["Unit B"], line = 'q')
plt.title('Unit B')
plt.show()
```

Comparing Evidence using t-statistics


```python
statistic , p_value = stats.ttest_ind(df['Unit A'],df['Unit B'], alternative = 'two-sided')
print('p_value=',p_value)
```
